```python
from scipy import stats
import numpy as np

A = np.array([78,84,92,88,75,80,85,90,87,7978,84,92,88,75,80,85,90,87,79])
B = np.array([82,88,75,90,78,85,88,77,92,8082,88,75,90,78,85,88,77,92,80])

t_val, p_val = stats.ttest_ind(A, B)

alpha = 0.05
df = len(A)+len(B)-2

crit_t = stats.t.ppf(1 - alpha/2, df)

print("T-value:", t_val)
print("P-Value:", p_val)
print("Critical t-value:", crit_t)

print('T-test Result:')
if np.abs(t_val) >crit_t:
    print('Significant difference found.')
else:
    print('No significant difference.')

print('P-test Result:')
if p_val >alpha:
    print('Fail to reject H0. No strong evidence of difference.')
else:
    print('Reject H0. Significant difference found.')
```

```python
from scipy import stats
import numpy as np

A = np.array([4, 4, 7, 16, 20, 11, 13, 9, 11, 15])
B = np.array([15, 16, 14, 14, 22, 22, 23, 18, 18, 19])

t_val, p_val = stats.ttest_rel(A, B)

alpha = 0.05
df = len(A)-1

c_t = stats.t.ppf(1 - alpha/2, df)

print("T-value:", t_val)
print("P-Value:", p_val)
print("Critical t-value:", c_t)

print('T-test:')
if np.abs(t_val) >c_t:
    print('Significant difference found.')
else:
    print('No significant difference.')

print('P-test:')
if p_val >alpha:
    print('Reject H0')
else:
    print('Fail to reject H0')
```

T-value (-4.95) is less than the negative critical t-value (-2.2622), so we reject the null hypothesis.
P-value (0.00079) is less than 0.05, indicating significance.
There is a significant difference between Math1 and Math2 scores. The difference is unlikely due to chance.

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Convert terminationdate_key to datetime and then to ordinal for plotting/regression
data['terminationdate_key_dt'] = pd.to_datetime(data['terminationdate_key'], errors='coerce')
data['terminationdate_key_ord'] = data['terminationdate_key_dt'].map(lambda x: x.toordinal() if pd.notnull(x) else np.nan)

fig, ax = plt.subplots()
ax.plot(data['terminationdate_key_dt'], data['length_of_service'], color='0.75')
ax = sns.regplot(x='terminationdate_key_ord', y='length_of_service', data=data, ci=None, scatter_kws=dict(color='0.25'))
ax.set_title('Time Plot of length_of_service')
```

```python
x=data['age']
y=data["length_of_service"]
x_train,x_test,y_train,y_test=train_test_split(x.values.reshape(-1,1),y,test_size=0.1,random_state=42)
model = LinearRegression()
model.fit(x_train, y_train) 
model.score(x_test,y_test)
sns.kdeplot(data['age'],shade=True,color="blue")
```
